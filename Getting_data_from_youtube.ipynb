```python
import csv
import googleapiclient.discovery
import googleapiclient.errors

def get_youtube_comments(video_id, developer_key):
    # Build the YouTube API service
    youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=developer_key)

    # Make a request to get the comments for the specified video
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=500
    )
    response = request.execute()

    # Extract comments from the response
    comments = []
    for item in response.get('items', []):
        comment_text = item['snippet']['topLevelComment']['snippet']['textDisplay']
        comments.append(comment_text)

    return comments

def save_comments_to_csv(video_id, developer_key):
    # Get comments for the specified video
    comments = get_youtube_comments(video_id, developer_key)

    # Create a CSV file and write comments to it
    #csv_file_name = f"{video_id}_comments.csv"
    csv_file_name = "padma_bridge_1.csv"
    with open(csv_file_name, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(['Comment'])

        for comment in comments:
            csv_writer.writerow([comment])

    print(f"Comments have been saved to {csv_file_name}")

# Example usage:
video_id = "1BN1hroJMQ"
DEVELOPER_KEY = "AIzaSyDy1OS0k7ZeTqHAdUEBGfovbuhJzL9n6js"
save_comments_to_csv(video_id, DEVELOPER_KEY)
```

# Making function that will take multiple video id and put into one single csv file

```python
import csv
import googleapiclient.discovery
import googleapiclient.errors

def get_youtube_comments(video_id, developer_key):
    # Build the YouTube API service
    youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=developer_key)

    # Make a request to get the comments for the specified video
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=1000
    )
    response = request.execute()

    # Extract comments from the response
    comments = []
    for item in response.get('items', []):
        comment_text = item['snippet']['topLevelComment']['snippet']['textDisplay']
        comments.append(comment_text)

    return comments

def save_comments_to_csv(video_ids, developer_key, output_csv):
    # Create a CSV file and write comments to it
    with open(output_csv, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(['Comment'])

        # Iterate through each video ID and get comments
        for video_id in video_ids:
            comments = get_youtube_comments(video_id, developer_key)

            # Write comments to the CSV file
            for comment in comments:
                csv_writer.writerow([comment])

    print(f"Comments have been saved to {output_csv}")

# Example usage with multiple video IDs:
video_ids = [ "z_atOd7S3xU" ,"iAcowXEDVKA" ,"g5FZkvo1J68"  ]
DEVELOPER_KEY = "AIzaSyAeVwggAZ-Z3CweUDWve2VbXxKSli3vnSs"
output_csv = "main_pandma_bridge_4.csv"

save_comments_to_csv(video_ids, DEVELOPER_KEY, output_csv)
```

# Adding multiple csv file to one csv file [ function ]

```python
import pandas as pd

def merge_csv_files(file_paths, output_path):
    # Initialize an empty DataFrame to store the combined data
    combined_data = pd.DataFrame()

    # Loop through each file path and read the CSV data
    for path in file_paths:
        data = pd.read_csv(path)
        combined_data = combined_data.append(data, ignore_index=True)

    # Write the combined data to a new CSV file
    combined_data.to_csv(output_path, index=False)

# Example usage:
file_paths = ['/content/main_pandma_bridge.csv', '/content/main_pandma_bridge_1.csv', '/content/main_pandma_bridge_3.csv', '/content/main_pandma_bridge_4.csv']
output_path = '/content/Padma_Bridge_DataSet_comment.csv'

merge_csv_files(file_paths, output_path)
```
